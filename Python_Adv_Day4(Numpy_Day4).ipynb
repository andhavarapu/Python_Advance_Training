```python
# Dot function (or) method

A = np.array([[1,1],[0,1]])
B = np.array([[2,0],[3,4]])

print ('A:\n', A)
print ('B:\n', B)
```

```python
A*B # Element wise multiplication happen
```

```python
# Modifying an existing array rather than creating a new one.

a = np.array ([10,10,10])
b = np.array ([5,5,5])
a +=3
print(a)
b +=a
print(b)
```

```python
# Universal Functions : Trignometry, Arthamatic, Complex and Statistical Functions are called as Universal functions,
# Trignometric Functions:

angles = np.array([0,30,45,60,90])
# Angles need to be converted to radians by multiplying by pi/180
angles_radians = angles*np.pi/180
print (angles_radians)
print('SINE of angles in the array:')
print(np.sin(angles_radians))

print('\nTAN of angles in the array:')
print(np.tan(angles_radians))

print('\nCOS of angles in the array:')
print(np.cos(angles_radians))
```

```python
# Statistical Functions:
test_scores = np.array([32.32, 56.98, 21.52, 44.32, 55.63])
print('\nMean Test Scores of the students:')
print(np.mean(test_scores))

print('\nMedian Test Scores of the students:')
print(np.median(test_scores))
```
